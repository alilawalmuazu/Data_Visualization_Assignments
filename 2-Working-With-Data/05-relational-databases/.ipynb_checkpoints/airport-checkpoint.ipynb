```python
import sqlite3
```

```python
conn = sqlite3.connect('airports.db')
```

```python
# Create a cursor object
cursor = conn.cursor()

# Query 1: All city names in the Cities table
cursor.execute("SELECT city FROM Cities;")
all_city_names = cursor.fetchall()
print("All city names in the Cities table:")
for city in all_city_names:
    print(city[0])
```

```python

# Query 2: All cities in Ireland in the Cities table
cursor.execute("SELECT city FROM Cities WHERE country = 'Ireland';")
all_cities_in_ireland = cursor.fetchall()
print("\nAll cities in Ireland in the Cities table:")
for city in all_cities_in_ireland:
    print(city[0])
```

```python
# Query 3: All airport names with their city and country
cursor.execute("""
    SELECT airports.name AS airport_name, cities.city AS city, cities.country AS country
    FROM airports
    JOIN cities ON airports.city_id = cities.id;
""")
all_airport_info = cursor.fetchall()
print("\nAll airport names with their city and country:")
for airport in all_airport_info:
    print(airport[0], "-", airport[1], ",", airport[2])
```

```python
# Query 4: All airports in London, United Kingdom
cursor.execute("""
    SELECT airports.name AS airport_name, cities.city AS city, cities.country AS country
    FROM airports
    JOIN cities ON airports.city_id = cities.id
    WHERE cities.city = 'London' AND cities.country = 'United Kingdom';
""")
airports_in_london_uk = cursor.fetchall()
print("\nAll airports in London, United Kingdom:")
for airport in airports_in_london_uk:
    print(airport[0], "-", airport[1], ",", airport[2])
```

```python
# Close the connection
conn.close()
```
